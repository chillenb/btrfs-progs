```python
import numpy as np
```

```python
with open('extentsmap.txt', 'r') as f:
    lines = f.readlines()
extents = []
blocks = []
for l in lines:
    e, b = l.split()
    extents.append(int(e,16))
    blocks.append(int(b,16))
bl = np.array(blocks)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ex = np.array(extents)
```

```python
plt.scatter(ex,bl,marker='+')
```

```python
import intervaltree
```

```python
etree = intervaltree.IntervalTree()
```

```python
for ext, blk in zip(extents, blocks):
    etree.addi(ext, ext+2**30, blk)
    
```

```python
etree.overlaps(extents[0])
```
